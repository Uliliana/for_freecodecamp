```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('adult.data.csv')
df.head(10)
```

```python
race_count = df['race'].value_counts()
race_count
```

```python
average_age = df.groupby('sex').mean()
average_age
```

```python
average_age_men=average_age['age'].iloc[1]
average_age_men
```

```python
education = df['education'].value_counts()
count_education = len(df['education'])
count_education
```

```python
count_bachelors = len(df[df['education'] == 'Bachelors'])
count_bachelors
```

```python
# What is the percentage of people who have a Bachelor's degree?
percentage_bachelors = count_bachelors/count_education*100
percentage_bachelors
```

```python
# What percentage of people with advanced education (`Bachelors`, `Masters`, or `Doctorate`) make more than 50K?
advanced_education =  df[(df['education'] == 'Bachelors')| (df['education'] =='Masters') | (df['education'] =='Doctorate')]
advanced_education
```

```python
salary_more_50 =df[df['salary']=='>50K']
salary_more_50
```

```python
count_salary_more_50 = len(salary_more_50.value_counts())
count_salary_more_50
```

```python
education_plus_salary = pd.merge(advanced_education, salary_more_50, how = 'inner')
education_plus_salary
```

```python
# What percentage of people with advanced education (`Bachelors`, `Masters`, or `Doctorate`) make more than 50K?
count_education_plus_salary = len(education_plus_salary.value_counts())
count_education_plus_salary
```

```python
df_count = len(df.value_counts())
df_count
```

```python
higher_education_rich = count_education_plus_salary/df_count*100
higher_education_rich
```

```python
without_adv_education = df[(df['education'] != 'Bachelors') & (df['education'] !='Masters') & (df['education'] !='Doctorate')]
without_adv_education
without_adv_education_1 = len(without_adv_education.value_counts())
without_adv_education_1
```

```python
#What percentage of people without advanced education make more than 50K?
without_adv_education_count = df_count - len(advanced_education.value_counts())
without_adv_education_count
```

```python
without_adv_education_plus_salary = pd.merge(without_adv_education, salary_more_50, how = 'inner')
without_adv_education_plus_salary
count_without_adv_education_plus_salary = len(without_adv_education_plus_salary.value_counts())
count_without_adv_education_plus_salary
```

```python
lower_education_rich = count_without_adv_education_plus_salary/df_count*100
lower_education_rich
```

```python
#What is the minimum number of hours a person works per week?
min_work_hours = min(df['hours-per-week'], key=lambda i: int(i))
min_work_hours
```

```python
#What percentage of the people who work the minimum number of hours per week have a salary of more than 50K?
df_num_min_workers = df[(df['hours-per-week'] == 1 ) & (df['salary']=='>50K')]
num_min_workers = len(df_num_min_workers.value_counts())
num_min_workers
```

```python
df_num_min_workers
```

```python
percentage_3 = num_min_workers/df_count*100
percentage_3
```

```python
#What country has the highest percentage of people that earn >50K and what is that percentage?
country_count = df['native-country'].value_counts()
country_count
```

```python
highest_earning_country = 'United-States'
```

```python

highest_earning_country_1 = len(df[(df['salary']=='>50K')&(df['native-country'] == 'United-States')].value_counts())
highest_earning_country_2 = len(df[df['native-country'] == 'United-States'].value_counts())
highest_earning_country_percentage =  highest_earning_country_1/highest_earning_country_2*100
highest_earning_country_percentage
```

```python
#Identify the most popular occupation for those who earn >50K in India.
pop_occupation_in_India = df[(df['salary']=='>50K') & (df['native-country'] == 'India')]
top_IN_occupation = pop_occupation_in_India['occupation'].value_counts()
top_IN_occupation = top_IN_occupation.idxmax()
```
